```python
import pandas as pd
```

```python
# IMDb data
imdb_df = pd.read_csv('IMDB-Movie-Data.csv')

# Rotten Tomatoes data
rt_df = pd.read_csv('rotten_tomatoes_movies.csv')
```

```python
imdb_df.head()
```

```python
rt_df.head()
```

```python
print(imdb_df.columns)
print(rt_df.columns)
```

```python
# Rename the relevant columns in the Rotten Tomatoes dataset
rt_df.rename(columns={
    'movie_title': 'Title',           # Rename 'movie_title' to 'Title'
    'tomatometer_rating': 'Rating',   # Rename 'tomatometer_rating' to 'Rating'
    'original_release_date': 'Year',  # Rename 'original_release_date' to 'Year'
    'genres':'Genre'                  # Rename 'genres' to 'genre'
}, inplace=True)
```

```python
print(rt_df.columns)
```

```python
rt_df['Year'] = rt_df['Year'].str[:4]
```

```python
rt_df.head()
```

```python
# Extract relevant columns from IMDb data
imdb_selected = imdb_df[['Title', 'Genre', 'Rating', 'Year']]

# Extract relevant columns from Rotten Tomatoes data
rt_selected = rt_df[['Title', 'Genre', 'Rating', 'Year']]
```

```python
# Convert 'Year' to string in both datasets
imdb_selected['Year'] = imdb_selected['Year'].astype(str)
rt_selected['Year'] = rt_selected['Year'].astype(str)

# Now merge the two datasets
merged_df = pd.merge(imdb_selected, rt_selected, on=['Title', 'Year'], how='inner')

# Check the merged dataframe
print(merged_df.head())
```

```python
# Keep IMDb genre and drop the duplicate from Rotten Tomatoes (Genre_y)
merged_df.drop('Genre_y', axis=1, inplace=True)
merged_df.rename(columns={'Genre_x': 'Genre'}, inplace=True)
```

```python
merged_df.head()
```

```python
# Rename 'Rating_x' to 'Rating_IMDB' and 'Rating_y' to 'Rating_RT'
merged_df.rename(columns={'Rating_x': 'Rating_IMDB', 'Rating_y': 'Rating_RT'}, inplace=True)
merged_df.head()
```
